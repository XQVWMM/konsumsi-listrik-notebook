```python
import numpy as np 
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sns

data_2021 = pd.read_csv('/Users/Jeff/Documents/ML/DATASET/2021/Hasil_2021.csv')
data_2022 = pd.read_csv('/Users/Jeff/Documents/ML/DATASET/2022/Hasil_2022.csv')
data_2023 = pd.read_csv('/Users/Jeff/Documents/ML/DATASET/2023/Hasil_2023.csv')
data_JP = pd.read_csv('/Users/Jeff/Documents/ML/DATASET/Jumlah penduduk/Hasil_JP.csv')
data_KP = pd.read_csv('/Users/Jeff/Documents/ML/DATASET/Kapasitas Pembangkit/Hasil_KP.csv')
data_final = pd.read_csv('Hasil_Gabungan.csv')
```

```python
features = ['Residential_2021', 'Business_2021', 'Industrial_2021', 'Social_2021', 'Gov_Office_2021', 'Pub_Street_2021', 'Total_2021', 'JP_2021', 'KP_2021']

data_T2021 = data_final[features]

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_T2021[features])

kmeans_2021 = KMeans(n_clusters=2, random_state=42)
data_T2021['Cluster'] = kmeans_2021.fit_predict(data_scaled)

centroid_2021 = kmeans_2021.cluster_centers_
print("Centorid_2021 : ")
print(centroid_2021)
```

```python

data_final['Cluster'] = kmeans_2021.fit_predict(data_scaled)

provinsi_per_cluster = data_final[['Province', 'Cluster']].sort_values(by='Cluster')

for cluster in range(2):  
    print(f"Provinsi dalam Cluster {cluster}:")
    print(provinsi_per_cluster[provinsi_per_cluster['Cluster'] == cluster]['Province'].tolist())
    print("\n")

```

```python
features = ['Residential_2022', 'Business_2022', 'Industrial_2022', 'Social_2022', 'Gov_Office_2022', 'Pub_Street_2022', 'Total_2022', 'JP_2022', 'KP_2022']

data_T2022 = data_final[features]

scaler_2022 = StandardScaler()
data_scaled_2022 = scaler_2022.fit_transform(data_T2022[features])

kmeans_2022 = KMeans(n_clusters=2, random_state=42)
data_T2022['Cluster'] = kmeans_2022.fit_predict(data_scaled)

centroid_2022 = kmeans_2022.cluster_centers_
print("Centorid_2022 : ")
print(centroid_2022)
```

```python
features = ['Residential', 'Business', 'Industrial', 'Social', 'Gov_Office', 'Pub_Street', 'Total', 'JP_2023', 'KP_2023']

data_T2023 = data_final[features]

scaler_2023 = StandardScaler()
data_scaled_2023 = scaler_2023.fit_transform(data_T2023[features])

kmeans_2023 = KMeans(n_clusters=2, random_state=42)
data_T2023['Cluster'] = kmeans_2023.fit_predict(data_scaled)

centroid_2023 = kmeans_2023.cluster_centers_
print("Centorid_2023 : ")
print(centroid_2023)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(features, centroid_2021[0], label="Centroid 1 - 2021", marker='o', color='black', linestyle='-', linewidth=2)
plt.plot(features, centroid_2022[0], label="Centroid 1 - 2022", marker='o', color='yellow', linestyle='-', linewidth=2)
plt.plot(features, centroid_2023[0], label="Centroid 1 - 2023", marker='o', color='pink', linestyle='-', linewidth=2)

plt.xlabel('Fitur')
plt.ylabel('Nilai Centroid')
plt.title('Perbandingan Centroid untuk Setiap Klaster (2021, 2022, 2023)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_score_2021 = silhouette_score(data_scaled, data_T2021['Cluster'])
print(f"Silhouette Score untuk 2021: {sil_score_2021}")

sil_score_2022 = silhouette_score(data_scaled_2022, data_T2022['Cluster'])
print(f"Silhouette Score untuk 2022: {sil_score_2022}")

sil_score_2023 = silhouette_score(data_scaled_2023, data_T2023['Cluster'])
print(f"Silhouette Score untuk 2023: {sil_score_2023}")
```
