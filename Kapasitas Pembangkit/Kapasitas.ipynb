```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("Kapasitas Pembangkit Listrik.csv")
data.columns = ['Province','Tahun_2021', 'Tahun_2022', 'Tahun_2023']
data = data.drop(index=[0,1])
data = data.iloc[:-1]

data['Tahun_2021'] = pd.to_numeric(data['Tahun_2021'], errors='coerce')
data['Tahun_2022'] = pd.to_numeric(data['Tahun_2022'], errors='coerce')
data['Tahun_2023'] = pd.to_numeric(data['Tahun_2023'], errors='coerce')
data = data.dropna()

data['Province'] = data['Province'].apply(lambda x: ' '.join([word.capitalize() for word in x.split()]))

provinsi_mapping = {
    "Dki Jakarta": "DKI Jakarta",
    "Di Yogyakarta": "DI Yogyakarta",
    "Kep. Riau": "Kep. Riau",
    "Jawa Barat": "Jawa Barat",
    "Sulawesi Tenggara": "Sulawesi Tenggara"
}

data['Province'] = data['Province'].replace(provinsi_mapping)
data.to_csv('Hasil_KP.csv', index=False)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(15, 18))  
features = ['Tahun_2021', 'Tahun_2022', 'Tahun_2023']

for i, feature in enumerate(features):
    ax = axes[i] 
    sns.barplot(x=data['Province'], y=data[feature], ax=ax)
    ax.set_title(f'{feature} per Provinsi')
    ax.set_xlabel('Province')
    ax.set_ylabel(feature)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)  

plt.tight_layout()
plt.show()
```
