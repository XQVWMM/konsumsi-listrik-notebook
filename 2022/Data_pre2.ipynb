```python
import pandas as pd
data = pd.read_csv("../2022/K2022.csv")

data.columns = ['Province', 'Residential', 'Industrial', 'Business', 'Social', 'Gov_Office', 'Pub_Street', 'Total', 'Percentage']
data = data.drop(index=[0,1])
data.head()
```

```python
uiw_kalimantan = data[data['Province'].str.contains('UIW Kalimantan Timur dan Utara', na=False)]

print(uiw_kalimantan) 
```

```python
def fix_numeric_format(value):
    if isinstance(value, str):
        value = value.replace('.', '') 
        value = value.replace(',', '.')  
    return value

numeric_columns = data.select_dtypes(include=['object']).columns  

for column in numeric_columns:
    if column != 'Province':  
        data[column] = data[column].apply(fix_numeric_format)  
        data[column] = pd.to_numeric(data[column], errors='coerce')  

```

```python
new_data = {
    'Residential': 2486.11,  
    'Industrial': 454.56,
    'Business': 1091.99,
    'Social': 212.35,
    'Gov_Office': 180.92,
    'Pub_Street': 53.38,
    'Total': 4479.30,
    'Percentage': 1.64
}
for key in new_data:
    new_data[key] = new_data[key] / 2

kalimantan_timur = new_data.copy()
kalimantan_timur['Province'] = 'Kalimantan Timur'

kalimantan_utara = new_data.copy()
kalimantan_utara['Province'] = 'Kalimantan Utara'
data = pd.concat([data, pd.DataFrame([kalimantan_timur, kalimantan_utara])], ignore_index=True)

print(data.tail())
```

```python
provinsi_valid = [
    "Aceh", "Sumatera Utara", "Sumatera Barat", "Riau", "Kep. Riau", "Jambi", 
    "Sumatera Selatan", "Kep. Bangka Belitung", "Bengkulu", "Lampung", "DKI Jakarta", 
    "Jawa Barat", "Banten", "Jawa Tengah", "DI Yogyakarta", "Jawa Timur", "Kalimantan Barat", 
    "Kalimantan Tengah", "Kalimantan Selatan", "Kalimantan Timur", "Kalimantan Utara", 
    "Sulawesi Utara", "Gorontalo", "Sulawesi Tengah", "Sulawesi Selatan", "Sulawesi Barat", 
    "Sulawesi Tenggara", "Bali", "Nusa Tenggara Barat", "Nusa Tenggara Timur", "Maluku", 
    "Maluku Utara", "Papua Barat", "Papua"
]
def clean_province_name(province_name):
    province_name = province_name.replace('-', '').strip()  
    province_name = province_name.replace("UIW ", "").strip()  
    province_name = province_name.replace("UID", "").strip()  
    province_name = province_name.replace("J a v a", "Indonesia").replace("I n d o n e s i a", "Indonesia")  
    province_name = province_name.replace("D.I. Yogyakarta", "DI Yogyakarta")
    province_name = province_name.replace("Kepulauan Riau", "Kep. Riau")
    province_name = province_name.replace("Bangka Belitung", "Kep. Bangka Belitung")
    province_name = province_name.replace("Jakarta Raya", "DKI Jakarta")

    return province_name

data['Province'] = data['Province'].apply(clean_province_name)
data_filtered_new = data[data['Province'].isin(provinsi_valid)]
data_filtered_new.to_csv('Hasil_2022.csv', index=False)
```
