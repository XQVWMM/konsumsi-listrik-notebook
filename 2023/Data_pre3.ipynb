```python
import pandas as pd
data = pd.read_csv("../2023/K2023.csv")

```

```python
data.columns = ['Province', 'Residential', 'Industrial', 'Business', 'Social', 'Gov_Office', 'Pub_Street', 'Total', 'Percentage']
data = data.drop(index=[0,1])
data.head()
```

```python
numeric_columns = data.select_dtypes(include=['object']).columns  

for column in numeric_columns:
    if column != 'Province':  
        data[column] = data[column].replace({',': ''}, regex=True)  
        data[column] = pd.to_numeric(data[column], errors='coerce') 

data.to_csv('data_no_comma.csv', index=False)
print(data.head())
```

```python
provinsi_valid = [
    "Aceh", "Sumatera Utara", "Sumatera Barat", "Riau", "Kep. Riau", "Jambi", 
    "Sumatera Selatan", "Kep. Bangka Belitung", "Bengkulu", "Lampung", "DKI Jakarta", 
    "Jawa Barat", "Banten", "Jawa Tengah", "DI Yogyakarta", "Jawa Timur", "Kalimantan Barat", 
    "Kalimantan Tengah", "Kalimantan Selatan", "Kalimantan Timur", "Kalimantan Utara", 
    "Sulawesi Utara", "Gorontalo", "Sulawesi Tengah", "Sulawesi Selatan", "Sulawesi Barat", 
    "Sulawesi Tenggara", "Bali", "Nusa Tenggara Barat", "Nusa Tenggara Timur", "Maluku", 
    "Maluku Utara", "Papua Barat", "Papua"
]
# Fungsi untuk membersihkan nama provinsi
def clean_province_name(province_name):
    if isinstance(province_name, str):  # Memastikan hanya memproses nilai string
        province_name = province_name.replace('-', '').strip()  
        province_name = province_name.replace("UIW ", "").strip()  
        province_name = province_name.replace("UID", "").strip()  
        province_name = province_name.replace("J a v a", "Indonesia").replace("I n d o n e s i a", "Indonesia")  
        province_name = province_name.replace("North Sumatera", "Sumatera Utara")
        province_name = province_name.replace("South Sumatera", "Sumatera Selatan")
        province_name = province_name.replace("Central Java", "Jawa Tengah")
        province_name = province_name.replace("West Java", "Jawa Barat")
        province_name = province_name.replace("D.I. Yogyakarta", "DI Yogyakarta")
        province_name = province_name.replace("West Kalimantan", "Kalimantan Barat")
        province_name = province_name.replace("East Java", "Jawa Timur")
        province_name = province_name.replace("West Papua", "Papua Barat")
        province_name = province_name.replace("Papua and West Papua", "Papua") 

        province_name = province_name.replace("West Sumatera", "Sumatera Barat")
        province_name = province_name.replace("Riau Archipelago", "Kep. Riau")
        province_name = province_name.replace("Bangka Belitung", "Kep. Bangka Belitung")
        province_name = province_name.replace("South Kalimantan", "Kalimantan Selatan")
        province_name = province_name.replace("Central Kalimantan", "Kalimantan Tengah")
        province_name = province_name.replace("East and North Kalimantan", "Kalimantan Timur")
        province_name = province_name.replace("North Sulawesi", "Sulawesi Utara")
        province_name = province_name.replace("Central Sulawesi", "Sulawesi Tengah")
        province_name = province_name.replace("South Sulawesi", "Sulawesi Selatan")
        province_name = province_name.replace("Southeast Sulawesi", "Sulawesi Tenggara")
        province_name = province_name.replace("West Sulawesi", "Sulawesi Barat")
        province_name = province_name.replace("North Maluku", "Maluku Utara")
        province_name = province_name.replace("West Nusa Tenggara", "Nusa Tenggara Barat")
        province_name = province_name.replace("East Nusa Tenggara", "Nusa Tenggara Timur")
        province_name = province_name.replace("Jakarta Raya", "DKI Jakarta")
    return province_name

data['Province'] = data['Province'].apply(clean_province_name)
data_filtered_new = data[data['Province'].isin(provinsi_valid)]
data_filtered_new.to_csv('data_filtered.csv', index=False)
print(data_filtered_new.head())
```

```python
kalimantan_timur = data_filtered_new[data_filtered_new['Province'] == 'Kalimantan Timur']
numeric_columns = ['Residential', 'Industrial', 'Business', 'Social', 'Gov_Office', 'Pub_Street', 'Total', 'Percentage']
for column in numeric_columns:
    kalimantan_timur[column] = pd.to_numeric(kalimantan_timur[column], errors='coerce') 

for column in numeric_columns:
    nilai_bagi = kalimantan_timur[column].values[0] / 2
    kalimantan_timur[column] = nilai_bagi 

data_filtered_new = data_filtered_new[data_filtered_new['Province'] != 'Kalimantan Timur']
kalimantan_timur['Province'] = 'Kalimantan Timur' 
data_filtered_new = pd.concat([data_filtered_new, kalimantan_timur], ignore_index=True)
kalimantan_utara = kalimantan_timur.copy() 
kalimantan_utara['Province'] = 'Kalimantan Utara' 
data_filtered_new = pd.concat([data_filtered_new, kalimantan_utara], ignore_index=True)

data_filtered_new = data_filtered_new[data_filtered_new['Province'].isin(provinsi_valid)]
data_filtered_new.to_csv('Hasil_2023.csv', index=False)
print(data_filtered_new.head())
```
